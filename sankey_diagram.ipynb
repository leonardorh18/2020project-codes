```python
def checkMissing(month, targets):
  complete = []
  for t in range(30):
    if calc_topic(month + 1, t) not in targets:
      complete.append(calc_topic(month+1, t))
  return complete
```

```python
def calc_weight(hots, topic):
  for idx, t in enumerate(hots):
    if t == topic:
      return idx+1
```

```python

ignore = []
sources = []
targets = []
lineColors = []
nodeColors = []
month = 1
aux = 1000
for i in range(1,23,2):
  for j in range(30):
    #if calc_topic(month, j) in ignore or month == 1:
      if type(mat[j][i]) == str and ',' in mat[j][i]:
          
          splt = mat[j][i].split(',')
          
          add = False
          for n in splt:

              sources.append(calc_topic(month, j) )
              targets.append(calc_topic(month + 1, int(n)) -1 )
              
              lineColors.append('rgb(200,200,200)')
              
              if calc_topic(month, j) in hots[month-1] and not add:
                w = calc_weight(hots[month-1], calc_topic(month, j))
                colorValue = int(255*w/10)-20
                nodeColors.append('rgb(255,{}, 0)'.format(colorValue))
                add = True
              elif not add:
                nodeColors.append('rgba(0,255,0, 0.3)')
                add = True

      else:
          if int(mat[j][i-1]) < 0:
            lineColors.append('rgba(255,255,255,0)')
          else:
            lineColors.append('rgb(200,200,200)')

          targets.append( calc_topic(month + 1, int(mat[j][i])) -1 )
          
          sources.append(calc_topic(month, j) )
          
          if calc_topic(month, j) in hots[month-1]:
            w = calc_weight(hots[month-1], calc_topic(month, j))
            colorValue = int(255*w/10)-20
            nodeColors.append('rgb(255,{}, 0)'.format(colorValue))
          else:
            nodeColors.append('rgba(0,255,0, 0.3)')

  complete = checkMissing(month, targets)
  node = 1000 + month
  nodeAux = 1000 + month+1

  for idx, c in enumerate(complete):
    lineColors.append('rgba(255,255,255,0)')
    sources.append(node)
    targets.append(c)


  sources.append(node)
  targets.append(nodeAux)
  lineColors.append('rgba(255,255,255,0)')
  
  #print(month)
  month += 1

```
